```python
#Extract file from the zip file using libraries
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #Extract file to dir
zip_ref.close() #close file
```

```python
#Read data from the file and handle NULL values
import pandas as pd
df = pd.read_csv('orders.csv', na_values= ['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#Rename all column names to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ','_')
df.head()
```

```python
#Drive new columns discount, sale price and profit
df['discount'] = df['list_price'] * df['discount_percent']/100
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['cost_price'] - df['sale_price']
df.head(20)
```

```python
#Convert order_date to DateTime format
df['order_date'] = pd.to_datetime(df['order_date'], format = '%Y-%m-%d')
df.info()
```

```python
#Drop cost price list price and discount percent columns
df.drop(columns=['list_price', 'discount_percent', 'cost_price'], inplace= True)
```

```python
#Importing mySQL libraries
import pymysql #this will work as adapter
from sqlalchemy import create_engine
```

```python
# Establishing a connection to the MySQL 'sales' database using SQLAlchemy
# and handling any connection errors that may occur
engine_mysql = create_engine("mysql+pymysql://root:root@localhost:3306/sales")
try:
    engine_mysql
    print("connected Successful")
except:
    print("Unable to connect")
```

```python
#Exporting my data to mySQL and if the data existes I am appending it
df.to_sql(name='orders', con=engine_mysql, if_exists='append', index=False)
```
